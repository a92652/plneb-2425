### Processamento do Texto

```python
import gensim

f1 = open("Harry_Potter_e_A_Pedra_Filosofal.txt", encoding = "utf-8")
f2 = open("Harry_Potter_Camara_Secreta-br.txt", encoding = "utf-8")

# é preciso ter uma lista de listas de tokens
sentences = []

for line in f1:
    tokens = list(gensim.utils.tokenize(line, lower = True))
    sentences.append(tokens)

for line in f2:
    tokens = list(gensim.utils.tokenize(line, lower = True))
    sentences.append(tokens)
```

```python
print(sentences[:10])
```

### Criar o Modelo

```python
from gensim.models import Word2Vec

model = Word2Vec(sentences, vector_size = 300, epochs = 20)
```

### Top 10 Palavras mais Comuns

```python
model.wv.key_to_index  # dicionário com as palavras conhecidas
model.wv.index_to_key[:10]  # top 10 palavras mais frequentes
# muitas stop words
```

```python
# sem stop words

#stopwords = ['a', 'o', 'e', 'de', 'do', 'da', 'em', 'que', 'com', 'para', 'se', 'um', 'uma', 'os']

import nltk
from nltk.corpus import stopwords

nltk.download('stopwords') 
stopwords_pt = set(stopwords.words('portuguese'))
```

```python
top_10 = model.wv.index_to_key[:100]

top_10_sem_stopwords = [w for w in top_10 if w not in stopwords_pt]
print(top_10_sem_stopwords[:10])
```

### Representação recorrendo a PCA

```python
from gensim.models import KeyedVectors
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

model = KeyedVectors.load_word2vec_format('model_harry.txt', binary=False) # carregar os vetores guardados

# escolher as palavras a visualizar
palavras = ['harry', 'rony', 'hermione', 'dumbledore', 'voldemort', 
            'draco', 'hogwarts', 'trouxas', 'grifinória', 'sonserina']
vetores = [model[word] for word in palavras]

# pca
pca = PCA(n_components=2)
vetores_2d = pca.fit_transform(vetores)

# gráfico
plt.figure(figsize=(10, 6))
plt.scatter(vetores_2d[:, 0], vetores_2d[:, 1], color='purple')

# Adicionar rótulos
for i, palavra in enumerate(palavras):
    plt.annotate(palavra, xy=(vetores_2d[i, 0], vetores_2d[i, 1]), fontsize=12)

plt.title('Representação Universo Harry Potter')
plt.grid(True)
plt.show()
```
